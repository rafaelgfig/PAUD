# Algorítmo para criar uma árvore de natal
   Inspirado por uma árvore de natal de uma hamburgueria onde era composta por tampas de cervejas Heineken, Budweiser e Stella Artois.
   
   Minha ideia era "contar" quantas tampinhas seriam necessárias para criar a minha árvore, daí surgiu esse programinha.

```python
# -*- coding: utf-8 -*-
# Created on Sat Feb 24 16:08:34 2018
# Cores
red   = '\033[31m'           # Cor Vermelha em ANSI
green = '\033[32m'           # Cor Verde em ANSI
white = '\033[37m'           # Cor Branca em ANSI
end   = '\033[0;0m'          # Final da saída de Cor

# Função para calcular os asteriscos verdes
def numGreen(i,cor,verdes): 
    if cor == green:         # Se o * for verde
        verdes += i          # Soma o número de *
    return(verdes)           # Retorna o número de *

# Função para calcular os asteriscos brancos
def numWhite(i,cor,brancos):
    if cor == white:         # Se o * for verde
        brancos += i         # Soma o número de *
    return(brancos)          # Retorna o número de *

# Função para calcular os asteriscos vermelhos
def numRed(i,cor,vermelhos):
    if cor == red:           # Se o * for verde
        vermelhos += i       # Soma o número de *
    return(vermelhos)        # Retorna o número de *

# Função para definir a cor da camada
def defCor(i,size):
    if i == (size - 2):      # Se for a penúltima, branca
        cor = white          # Definindo na cor branca
    elif i == (size - 1):    # Se for a última, cor vermelha
        cor = red            # Definindo na cor vermelha
    else:                    # Nas demais, cor verde
        cor = green          # Definindo cor verde
    return(cor)              # Retornando a cor

# Função para montar a árvore
def tree(margin, first1, size1, first2, size2, first3, size3, first4, size4):
    verdes = brancos = vermelhos = 0            # Declarando a quantidade de *
    last = size4 + first4                       # Declarando o tamanho da última camada
    for j in range(1,5):                        # Laço de repetição para as 4 partes da árvore
        if j == 1:
            first = first1                      # Definindo com quantos * começa a primeira
            size = first + size1                # Quantas camadas terá a primeira
        elif j == 2:
            first = first2                      # Definindo com quantos * começa a primeira
            size = first + size2                # Quantas camadas terá a segunda
        elif j == 3:
            first = first3                      # Definindo com quantos * começa a primeira
            size = first + size3                # Quantas camadas terá a terceira
        else:
            first = first4                      # Definindo com quantos * começa a primeira
            size = first + size4                # Quantas camadas terá a quarta
        for i in range(first,size):             # Criação das partes da árvore
            cor = defCor(i,size)                # Definindo a cor da camada
            print(cor + str((' ' * ((last - size) + margin + size - i)) + " *" * (i) + end))
            verdes = numGreen(i,cor,verdes)     # Somatório dos * verdes
            brancos = numWhite(i,cor,brancos)   # Somatório dos * brancos
            vermelhos = numRed(i,cor,vermelhos) # Somatório dos * vermelhos
    print("\nPara uma árvore de Natal feita em Python e com asteriscos, é preciso:\n"
          + str(verdes) + " Asteriscos " + green + "Verdes" + end + ",\n0"
          + str(vermelhos) + " Asteriscos " + red + "Vermelhos" + end + ",\n0"
          + str(brancos) + " Asteriscos " + white + "Brancos" + end + ",\n"
          + "120 Minutos de coding e \n001 Programador desocupado...")

# Montagem com o layout que eu escolhi
tree(5, 1,6, 4,6, 7,8, 11,8)
```
